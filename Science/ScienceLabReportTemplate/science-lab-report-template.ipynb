### Data Table

```python
# Change, add, or remove variable names below, then run this cell to create your data table
columns = ['Variable1','Variable2','Variable3']

import pandas as pd
import ipydatagrid
import ipywidgets as ipw
import plotly.express as px

def add_row(x):
    grid = dfq.get_visible_data()
    extra_row = [0] * len(grid.columns)
    grid.loc[len(grid.index)] = extra_row

    dfq.data = grid

def remove_row(x):
    grid = dfq.get_visible_data()
    dfq.data = grid.drop(grid.tail(1).index)
    
dfb = pd.DataFrame(columns=columns)
dfb.loc[len(dfb.index)] = [0]*len(columns) # ajouter une ligne avec des valeurs par d√©faut
dfq = ipydatagrid.DataGrid(dataframe=dfb, show_toolbar=True,editable=True,layout={'height':'200px','width':'500px'})

button_add_row = ipw.Button(description="Add Row", 
                            style=ipw.ButtonStyle(button_color='darkgreen'))
button_remove_row = ipw.Button(description="Remove Row", 
                               style=ipw.ButtonStyle(button_color='darkred'))

button_add_row.on_click(add_row)
button_remove_row.on_click(remove_row)

ipw.VBox([
    ipw.HBox([button_add_row,button_remove_row]),
    dfq
])
```
