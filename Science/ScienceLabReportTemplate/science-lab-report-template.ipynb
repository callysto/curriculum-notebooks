### Data Table

Change, add, or remove variable names in the first line of the code cell below, then run the cell to create your data table.

```python
columns = ['Variable1', 'Variable2', 'Variable3']

import pandas as pd
import ipydatagrid
import ipywidgets as ipw
import plotly.express as px

def add_row(x):
    grid = datagrid.get_visible_data()
    extra_row = [0] * len(grid.columns)
    grid.loc[len(grid.index)] = extra_row
    datagrid.data = grid

def remove_row(x):
    grid = datagrid.get_visible_data()
    datagrid.data = grid.drop(grid.tail(1).index)
    
dfb = pd.DataFrame(columns=columns)
dfb.loc[len(dfb.index)] = [0]*len(columns) # add a line with default values
datagrid = ipydatagrid.DataGrid(dataframe=dfb, show_toolbar=True, editable=True, auto_fit_columns=True)

button_add_row = ipw.Button(description='Add Row', style=ipw.ButtonStyle(button_color='darkgreen'))
button_remove_row = ipw.Button(description='Remove Row', style=ipw.ButtonStyle(button_color='darkred'))
button_add_row.on_click(add_row)
button_remove_row.on_click(remove_row)
ipw.VBox([ipw.HBox([button_add_row, button_remove_row]), datagrid])
```

### Graph

```python
graph_title = ''
xaxis_column = 'Variable1'
yaxis_column = 'Variable2'

df = datagrid.get_visible_data()
px.scatter(df, x=xaxis_column, y=yaxis_column, title=graph_title)
```
