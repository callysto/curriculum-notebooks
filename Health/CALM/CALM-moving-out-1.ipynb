## Part 1 - Personal Profile

üìòComplete these questions based on what your hope of plan to do with your life. Once you have answered all of the questions in the code cell below, run the cell using the `Run` button on the tool bar and it will store your answers in a text file.

```python
import ipywidgets as widgets

def create_widget(type,description,value,max,step):
    match type:
        case 'Text':
            return widgets.Text(
                    value=value,
                    description=description,
                    disabled=False,
                    indent=False, 
                    layout= widgets.Layout(width='400px'), 
                    style=dict(description_width='initial'),
                )
        case 'TextArea':
            return widgets.Textarea(
                    value=value,
                    description=description,
                    disabled=False,
                    indent=False, 
                    layout= widgets.Layout(width='400px',height='100px'), 
                    style=dict(description_width='initial'),
                )
        case 'Checkbox':
            return widgets.Checkbox(
                    value=value,
                    description=description,
                    disabled=False,
                    indent=False,
                )
        case 'IntSlider':
            return widgets.IntSlider(
                    value=value,
                    min = 0,
                    max= max,
                    step = step,
                    description=description,
                    disabled=False,
                    layout= widgets.Layout(width='500px'),
                    style=dict(description_width='initial'),
                )
        case 'IntRangeSlider':
            return widgets.IntRangeSlider(
                    value=value,
                    min=0,
                    max=max,
                    step=step,
                    description=description,
                    disabled=False,
                    continuous_update=False,
                    layout= widgets.Layout(width='500px'),
                    style=dict(description_width='initial'),           
                )
        case _:
            return None
            
Name = create_widget('Text','Your Name:','Enter your name',0,0)
Moveout = create_widget('IntSlider','I will move out at the age of:',0,30,1)
Education_path = create_widget('Text','My Education Path:','Enter what you want to study',0,0)
Occupation = create_widget('Text','My Occupation:','Enter what you want to do for work',0,0)
Salary = create_widget('IntRangeSlider','My estimated monthly Salary:',[2000,5000],20000,1000)
Married = create_widget('Checkbox','I want to married in my twenties',False,0,0)
Children = create_widget('IntSlider','# of children I want:',0,10,1)
House = create_widget('Checkbox','I want to buy a house in my twenties',False,0,0)
Location = create_widget('Text','I will be living near/in:','Enter the city/town you want to live in',0,0)
Vehicle = create_widget('Text','Vehicle I want to drive:','Enter the vehicle you want to own',0,0)
Possessions = create_widget('TextArea','Things I will also own','My Possessions',0,0)
Travel = create_widget('Checkbox','I want to travel in my twenties',False,0,0)
Destinations = create_widget('TextArea','Places I want to travel to','Travel Destinations',0,0)
Bucket_list = create_widget('TextArea','Things I want to do in my twenties','Bucket list',0,0)

box = widgets.VBox([Name,Moveout,Education_path,Occupation,Salary,Married,Children,House,Location,Vehicle,Possessions,Travel,Destinations,Bucket_list])

display(box)
```

```python
with open('moving_out1.txt', 'a') as f:
    f.truncate(0)
    for widget in box.children: 
        value = str(widget.value)
        desc = str.upper(widget.description)
        f.write(f"{desc}\n  {value}\n\n\n\n")
```

```python
list_widgets = [
    widgets.VBox([Name,Moveout]),
    widgets.VBox([Education_path,Occupation,Salary]),
    widgets.VBox([Married,House,Children,Location]),
    widgets.VBox([Vehicle,Possessions,Travel,Destinations,Bucket_list]),            
]
accordion = widgets.Accordion(children=list_widgets,titles=('General Info','Career','Settling Down','Quality of Life'))
display(accordion)
```
