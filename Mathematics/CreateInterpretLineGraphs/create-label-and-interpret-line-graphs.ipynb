# Creating, Labelling, and Interpreting Line Graphs

Line graphs are useful for visualizing (seeing) patterns in data. [Wikipedia](https://en.wikipedia.org/wiki/Line_chart) defines them as "a type of chart which displays information as a series of data points called 'markers' connected by straight line segments."

We can make line graphs using Python using code libraries such as [Plotly](https://plotly.com/python). A library is code that someone else has written that you can import into your project.

We need to import the [Plotly Express](https://plotly.com/python/plotly-express) part of the library, tell it what data to graph, and maybe add some labels.

Select the cell below this one then click the `▶Run` button on the toolbar. You can also do this to run code in other cells in this notebook.

```python
import plotly.express as px
x = [1, 2, 3, 4]
y = [5, 6, 7, 8]
px.scatter(x=x, y=y)
```

On a graph, the horizontal axis is the *x*-axis, and the vertical axis is the *y*-axis. We created a list called `x` for the *x* values and a list called `y` for the *y* values. `px.scatter(x, y)` then created and displayed the graph.

We can also create a line graph from the same data using `px.line()` instead.

```python
px.line(x=x, y=y)
```

## Interpreting Graphs

Here is a graph of average daytime high temperature in Edmonton, Alberta by month (using data from [NOAA](https://www.ncdc.noaa.gov/)). 

```python
temperatures = [-7, -3, 2, 11, 18, 21, 23, 22, 17, 11, 1, -5]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

px.line(x=months, y=temperatures, title='Average Temperature in Edmonton by Month').update_xaxes(title='Month').update_yaxes(title='Temperature (°C)')
```

Here's another graph with data for Edmonton, Alberta and Honolulu, Hawaii. For this one we are going to use [Plotly Graph Objects](https://plotly.com/python/graph-objects).

```python
averageHighEdmonton = [-7, -3, 2, 11, 18, 21, 23, 22, 17, 11, 1, -5]
averageHighHonolulu = [27, 27, 28, 28, 29, 30, 31, 31, 31, 30, 29, 28]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(go.Scatter(x=months,y=averageHighEdmonton, name='Edmonton'))
fig.add_trace(go.Scatter(x=months,y=averageHighHonolulu, name='Honolulu'))
fig.update_xaxes(title_text='Month')
fig.update_yaxes(title_text='Temperature (°C)')
fig.update_layout(title='Average High Temperatures in Edmonton and Honolulu')
fig.show()
```
